```python
def integrate_over_time(arr,integration_time,avg=False):

    '''
    Integrate over time by summing over in chunks of data
    
    Parameters:
    -------------
    arr: 'array'
         One-dimensional array on which the integration needs to be computed
    integration_time: 'int'
         Should be multiples of 4.
         The data would be divided in chunks depending on the number of data points
         and the integration time requested.
         
    Returns:
    ------------
    arr_int: 'array'       
    '''
    chunk_no = len(arr)//(integration_time/4)
    if avg:
        arr_int = np.array([sum(arr)/len(arr) for arr in np.array_split(arr,chunk_no)])
    else:
        arr_int = np.array([sum(arr) for arr in np.array_split(arr,chunk_no)])
    return arr_int
    
def convert_seconds_datetime(sod_i,date_obs):
    ''' 
    Function takes in seconds of day and the date of the observation as argument and returns
    a datetime object in a specified format
    '''
    for fmt in ('%Y-%m-%dT%H:%M:%S.%f', '%Y-%m-%dT%H:%M:%S'):
        try:
            return datetime.strptime(date_obs,fmt)+timedelta(seconds=sod_i)
        except:
            pass

def parse_stix_L1(fits_filename,give_spec=False,give_ql=False,give_full=False,
                  give_detector_counts=False,give_pixels=False,**kwargs):

    '''
    A function to parse the STIX full light-curve files such.
    Each file has a structure as follows:
    (time,detectors,pixels,energy_bins). STIX has 32 detectors,
    12 pixels for each detectors and 32 energy bins for each those pixels.
    The energy bins range from 0-150 keV. 
    
    The function gives options to get either a QL lightcurve for easy understanding
    of the count data or the full light curve with which you can develop,
    the full spectrum. In addition you can also get the individual detector or pixel counts.

    Parameteres:
    ---------------
    fits_filename: 'str'
                    The full path of the STIX fits file.

    CONFIGURATION OPTIONS
    ---------------------
    give_spec: bool
                    To give the spectrogram counts - (time,32 energy bins)
    give_ql: bool
                    To give the QL counts - by binning it into preset energy levels
    give_full: bool
                    To get the full lightcurve information.
                    This is essentially same as that of the 'give_spec', except, now
                    you can specify the energy levels between which you want the counts
    give_detector_counts: bool
                    To get the full detector counts for a set of detectors or for a
                    band of energy
    give_pixels: bool
                    To get the counts from each pixels either for a set of detectors or
                    for a band of energy


    '''
    #check if the user has given more than one TRUE configuration option.
    #only one opiton is allowed to be given
    assert np.sum([give_spec,give_ql,give_full,give_detector_counts,give_pixels]) == 1
    
    #unpack the specific detectors and energy bands requested
    kwargs['config'] = config
    en1,en2 = config['energy_range']
    dets = np.asarray(config['detectors'])-1
    pix = np.asarray(config['pixels'])-1

    #open the fits file
    hdulist = fits.open(fits_filename)
    header = hdulist[0].header
    #parse time which is given as seconds of the day
    time_sod = hdulist[2].data.field('time')
    #parse counts - (time, 32 detectors, 12 pixels, 32 energy bins)
    counts = hdulist[2].data.field('counts')
    #create labels for plotting, spectrogram and spectrum use cases
    labels = [str(x)+'-'+str(y)+' keV' for x,y in zip(hdulist[3].data['e_low'],hdulist[3].data['e_high'])]
    edges = [e_low+(e_high-e_low) for chn,e_low,e_high in hdulist[3].data]
    mean_energy = [(e_high+e_low)/2 for chn,e_low,e_high in hdulist[3].data]
    edges.insert(0,0)
    
    #find the indexes corresponding to the energy band requested.
    energy_idx = np.where((hdulist[3].data['e_low']>=en1) & (hdulist[3].data['e_low']<=en2))[0]
    energy_idx_edge = np.insert(energy_idx,-1,energy_idx[-1]+1)

    # Count arr:[time,detector,pixel,energy]
    if give_full:
        # take determined energy indexes and then sum over it 
        # over the detector dimensions and then pixel dimensions
        counts = counts[:,:,:,energy_idx].sum(axis=1).sum(axis=1)
    if give_ql:
        # give QL light curves for a set of 5 fixes energy bands
        # 0-4 keV, 6-10 keV, 10-15 keV, 15-25 keV, 25-50 keV
        counts = counts.sum(axis=1).sum(axis=1)
        # manualy sum up the energy bins
        counts = np.vstack([counts[:,0:7].sum(axis=1),
                           counts[:,7:12].sum(axis=1),
                           counts[:,13:19].sum(axis=1),
                           counts[:,19:25].sum(axis=1),
                           counts[:,25:31].sum(axis=1)]).T      
    if give_spec:
        # similar to full light curves but with no option to pick the energy band
        counts = counts.sum(axis=1).sum(axis=1)
    if give_detector_counts:
        # give the detector counts for a set of detectors for a given energy band
        # first sum over the energy bins and then over the pixels.
        counts = counts[:,:,:,energy_idx].sum(axis=3)
        counts = counts[:,dets,:].sum(axis=2)
    if give_pixels:
        # give the pixel counts for a set of detectors for a given energy band
        counts = counts[:,:,:,energy_idx].sum(axis=3)
        counts = counts[:,dets,:].sum(axis=1)
        counts = counts[:,pix]
    map_func = partial(convert_seconds_datetime,                       #create time array
                       date_obs=hdulist[0].header['DATE_BEG'])
    time_arr = np.asarray(list(map(map_func,time_sod)))
     
    data = {'time':time_arr,'data':counts,'labels':np.asarray(labels)[energy_idx],
            'edges':np.asarray(edges)[energy_idx_edge],'mean_energy':np.asarray(mean_energy)[energy_idx]}
    return header,data

def plot_spectrogram(count_arr,time_arr,labels):
    '''
    Function produces a spectrogram plot.
    Takes in as input the count arr ( usually with time on x axis and counts on y axis),
    the time array to plot time, and the labels.
    '''
    fig = plt.figure(figsize=(10,5),facecolor='w')
    axes = fig.subplots(nrows=1,ncols=1,)
    spec = axes.imshow(np.log10(count_arr).T,aspect='auto',cmap=cm.coolwarm,origin='lower',
               extent = [plt_dates.date2num(time_arr[0]),
                        plt_dates.date2num(time_arr[-1]),
                        0,32])
    yticks_idxs = [0,3,5,8,10,13,15,18,20,25,31]
    yticks_lbs = [labels[idx] for idx in yticks_idxs]
    axes.set_yticks(yticks_idxs)
    axes.set_yticklabels(yticks_lbs)
    date_format = plt_dates.DateFormatter('%H:%M')
    axes.xaxis.set_major_formatter(date_format)
    cbar = plt.colorbar(spec)
    cbar.set_label('Counts')
    
    plt.show()
    return fig, axes
    
def create_spectrum(count_arr,time_arr,dt_,mean_energy):
    '''
    Function takes in as input the count arr for producing spectrogram
    and returns a spectrum with bins corresponding to the mean of the native energy binning.
    '''
    time_idx = np.where((time_arr >=dt_[0]) &              #find the indexes corresponding to the time range requested
                          (time_arr <= dt_[1]))[0]
    #find the total time duration which we will use to divide the counts with
    time_dur = (time_arr[time_idx[-1]]-time_arr[time_idx[0]]).total_seconds()
    return count_arr[time_idx,:].sum(axis=0)/mean_energy/time_dur    

def plot_data(hxr_dates,hxrCounts,hxr_idx,
              only_detector=False,only_ql=False,full_lc=False,only_pixels=False):

    fig = plt.figure(figsize=(9,3),facecolor='w')
    axes = fig.subplots(nrows=1,ncols=1)
    cm = plt.get_cmap('turbo')
    NUM_COLORS = stix_data['data'].shape[1]
    hxr_dates = plt_dates.date2num(stix_data['time'])

    labels_tag={
        only_detector:['detector '+str(ii) for ii in dets],
        full_lc: stix_data['labels'],
        only_pixels: ['pixel '+str(ii) for ii in pixs],
        only_ql: f' '
    }
    labels = labels_tag[True]
    if only_ql:
        l1, = axes.plot_date(hxr_dates[hxr_idx],hxrCounts[hxr_idx,0].ravel(),'-',label='4-10 keV')
        l2, = axes.plot_date(hxr_dates[hxr_idx],hxrCounts[hxr_idx,1].ravel(),'-',label='10-15 keV')
        l3, = axes.plot_date(hxr_dates[hxr_idx],hxrCounts[hxr_idx,2].ravel(),'-',label='15-25 keV')
        l4, = axes.plot_date(hxr_dates[hxr_idx],hxrCounts[hxr_idx,3].ravel(),'-',label='25-50 keV')

    if full_lc or only_detector or only_pixels:
        for chan in range(stix_data['data'].shape[1]):
            lines = axes.plot(hxr_dates[hxr_idx],hxrCounts[hxr_idx,chan],label=labels[chan])
            lines[0].set_color(cm(1.*chan/NUM_COLORS))

    fig_titles={
        only_detector:f'Detector count rates for energy range {energy_range[0]}-{energy_range[1]} keV \n Pixels {pixs}',
        only_ql:f'Quicklook light curves',
        full_lc:f'Count rates for energy range {energy_range[0]}-{energy_range[1]} keV \n Detectors {dets} \n Pixels {pixs}',
        only_pixels:f'Pixel count rates for energy range {energy_range[0]}-{energy_range[1]} keV \n Detectors {dets}'
    }
        
    #l5 = axes.plot_date(hxr_dates[atten1],np.ones_like(hxr_dates[atten1])*hxrCounts[hxr_idx,:].max()*1.05,'r')
    axes.minorticks_on()
    axes.yaxis.set_minor_locator(mticker.MultipleLocator(5))
    #axes.set_yscale("log")
    axes.set_ylabel('counts')
    axes.legend(bbox_to_anchor=(1.05, 1),ncol=3)
    axes.yaxis.grid(True, 'major')
    date_format = plt_dates.DateFormatter('%H:%M')
    axes.xaxis.set_major_formatter(date_format)
    axes.text(0.5, -0.2, datetime.date(stix_data['time'][0]).strftime("%d/%m/%Y"), ha='center', va='center',
            transform=axes.transAxes)
    axes.set_title(fig_titles[True])
    plt.show()
    return fig, axes

    
def plot_integrated_data(hxr_dates,hxrCounts,hxr_idx,int_time,
              only_detector=False,only_ql=False,full_lc=False,only_pixels=False):

    hxr_dates_int = integrate_over_time(hxr_dates[hxr_idx],int_time,avg=True)
    hxr_dates_integrated = np.asarray(plt_dates.num2date(hxr_dates_int))
    hxr_counts_int = np.apply_along_axis(integrate_over_time, 0, 
                                        hxrCounts[hxr_idx,:],integration_time=int_time,avg=True)
    fig = plt.figure(figsize=(9,3),facecolor='w')
    axes = fig.subplots(nrows=1,ncols=1,)
    cm = plt.get_cmap('turbo')
    NUM_COLORS = stix_data['data'].shape[1]

    labels_tag={
        only_detector:['detector '+str(ii) for ii in dets],
        full_lc: stix_data['labels'],
        only_pixels: ['pixel '+str(ii) for ii in pixs],
        only_ql: f' '
    }
    labels = labels_tag[True]

    if only_ql:
        l1,=axes.plot_date(hxr_dates_int,hxr_counts_int[:,0],'royalblue',label='4-10 keV')
        l2,=axes.plot_date(hxr_dates_int,hxr_counts_int[:,1],'sandybrown',label='10-15 keV')
        l3,=axes.plot_date(hxr_dates_int,hxr_counts_int[:,2],'darkolivegreen',label='15-25 keV')
        l4,=axes.plot_date(hxr_dates_int,hxr_counts_int[:,3],'orangered',label='25-50 keV')
        
    if full_lc or only_detector or only_pixels:
        for chan in range(stix_data['data'].shape[1]):
            lines = axes.plot(hxr_dates_int,hxr_counts_int[:,chan],label=labels[chan])
            lines[0].set_color(cm(1.*chan/NUM_COLORS))
    #l5 = axes.plot_date(hxr_dates[atten1],np.ones_like(hxr_dates[atten1])*hxrCounts[hxr_idx,:].max()*1.05,'r')

    fig_titles={
        only_detector:f'Integrated({int_time} s) Detector count rates for energy range {energy_range[0]}-{energy_range[1]} keV ',
        only_ql:f'Integrated({int_time} s) Quicklook light curves',
        full_lc:f'Integrated({int_time} s) Count rates for energy range {energy_range[0]}-{energy_range[1]} keV \n Detectors {dets} \n Pixels {pixs}',
        only_pixels:f'Integrated({int_time} s) Pixel count rates for energy range {energy_range[0]}-{energy_range[1]} keV \n Detectors {dets}'
    }
    axes.minorticks_on()
    axes.yaxis.set_minor_locator(mticker.MultipleLocator(5))
    axes.set_yscale("log")
    axes.set_ylabel('counts')
    axes.legend(loc='upper left') #bbox_to_anchor=(1.05, 1))
    axes.yaxis.grid(True, 'major')
    date_format = plt_dates.DateFormatter('%H:%M')
    axes.xaxis.set_major_formatter(date_format)
    axes.text(0.5, -0.2, datetime.date(stix_data['time'][0]).strftime("%d/%m/%Y"), ha='center', va='center',
            transform=axes.transAxes)
    axes.set_title(fig_titles[True])
    plt.show()
    return fig, axes
    
```
