# Demonstrating sunxspex

This notebook demonstrates the capabilities of the sunxspex package.

```python
import astropy.units as u
import matplotlib.pyplot as plt
import numpy as np
```

```python
#%matplotlib notebook
```

```python
# Define energy bins.
binsize = 0.01
energy_edges = np.arange(3, 20+binsize, binsize) * u.keV

# Define temperature
temperature = 8.6 * u.MK

# Define emission measure
emission_measure = 7.7e45/u.cm**3
```

## Thermal

### Thermal Emission -- f_vth Equivalent

```python
from sunxspex.thermal import thermal_emission
```

```python
thermal_flux = thermal_emission(energy_edges, temperature, emission_measure)
```

```python
plt.plot(energy_edges[:-1], thermal_flux)
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Flux [{thermal_flux.unit}]")
plt.title("Thermal Emission")
```

## Non-thermal

### Thick Target

```python
from sunxspex.emission import bremsstrahlung_thick_target
```

```python
# Define parameters of electron distribution.
photon_energies = ((energy_edges[1:] + energy_edges[:-1]) / 2).value
low_slope = 7.7
high_slope = low_slope
low_energy_cutoff = 6.9
break_energy = 500
high_energy_cutoff = break_energy
total_electron_flux = 2.1e35
```

```python
flux_thick = bremsstrahlung_thick_target(photon_energies, low_slope, break_energy, high_slope,
                                         low_energy_cutoff, high_energy_cutoff)
flux_thick = flux_thick * total_electron_flux * u.ph / u.s / u.keV / u.cm**2
```

```python
plt.plot(photon_energies, flux_thick, label="thick target")
plt.yscale("log")
plt.xscale("log")
plt.legend()
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Photon Flux [{flux_thick.unit}]")
```

## Combining Thermal & Non-thermal

```python
flux = thermal_flux + flux_thick
```

```python
plt.plot(photon_energies, flux, label="total")
plt.plot(photon_energies, thermal_flux, linestyle="--", label="thermal")
plt.plot(photon_energies, flux_thick, linestyle="-.", label="non-thermal")
plt.legend()
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Photon Flux [{flux.unit}]")
plt.ylim(min(flux.value)*0.5, max(flux.value)*2)
```

# Extra Examples

## Thermal

### Changing Input Parameters

#### Temperature

```python
flux_20MK = thermal_emission(energy_edges, 20*u.MK, emission_measure)
```

```python
plt.step(energy_edges[:-1], thermal_flux, label=f"{temperature}")
plt.step(energy_edges[:-1], flux_20MK, label="20 MK")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Flux [{thermal_flux.unit}]")
plt.legend()
```

#### Observer Distance

```python
# Define observer distance
venus_distance = (0.725 * u.AU).to(u.cm)
flux_venus = thermal_emission(energy_edges, temperature, emission_measure, observer_distance=venus_distance)
```

```python
plt.step(energy_edges[:-1], thermal_flux, label="Earth")
plt.step(energy_edges[:-1], flux_venus, label="Venus")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Flux [{thermal_flux.unit}]")
plt.legend()
```

#### Relative Abundances

```python
# Get flux when Iron is double nominal abundance.
relative_abundances = ((26, 2))
flux_Fe2 = thermal_emission(energy_edges, temperature, emission_measure, relative_abundances=relative_abundances)
```

```python
plt.step(energy_edges[:-1], thermal_flux, label="Fe 1")
plt.step(energy_edges[:-1], flux_Fe2, '--', label="Fe 2")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Flux [{thermal_flux.unit}]")
plt.legend()
```

#### Abundance Type

```python
# Get flux using hybrid abundances.
flux_hybrid = thermal_emission(energy_edges, temperature, emission_measure, abundance_type="sun_hybrid")
```

```python
plt.step(energy_edges[:-1], thermal_flux, label="coronal")
plt.step(energy_edges[:-1], flux_hybrid, label="hybrid")
plt.yscale("log")
plt.xscale("log")
plt.xlabel(f"Energy [{energy_edges.unit}]")
plt.ylabel(f"Flux [{thermal_flux.unit}]")
plt.legend()
```
